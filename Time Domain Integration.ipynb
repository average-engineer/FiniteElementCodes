## Function for Mises Iteration to compute the Eigenvectors of a dynamic System

### The algorithm is given as follows:
For the Eigenvalue Problem described by $$\textbf{Ab} = \lambda \textbf{b}$$
* Choose an initial eigenvector $b_{0}$
* Calculate $c_{0} = Ab_{0}$
* $b_{1} = \frac{c_{0}}{|c_{0}|}$
* Calculate $c_{1} = Ab_{1}$
* Repeat the process till the $|b_{i+1} - b_{i}| < \Delta$, where $\Delta$ is the required tolerance

## Example
 A 4th order undamped system is given:
 ![sys](Images/5.jpg)

```python
# Function for the Mises Iteration
def misesIteration(A,b0,n,delta):
    # A: System Matrix
    # b0: Initial Guess
    # n: order of the system
    # delta: Required Tolerance
    # 1st Iteration:
    c = A*b0
    c_norm = c.norm()
    b1 = c/c_norm
    count = 1 # Number of Iterations
    while abs(b1 - b0)[0] > delta and abs(b1-b0)[-1] > delta:
        # Preparing the next iteration
        b0 = b1
        c = A*b1
        c_norm = c.norm()
        b1 = c/c_norm
        count = count + 1
    return b1,count
```

```python
# Mass Matrix
M = Matrix([[2,0],[0,1]])
# Stiffness Matrix
K = Matrix([[6,-2],[-2,4]])
# Damping Matrix
C = zeros(K.shape[0])
# Force/Load Vector
F = Matrix([[0],[10]]) # Constant Load Vector
# Initial Eigenvector guess
b0 = Matrix([[1],[1]])
# Required Tolerance
delta = 0.00001
n = 2 # DOF of the system
aa = ones(3,1)
aa[1] = -2
abs(aa)
eigvec1,noIter1 = misesIteration((M**-1)*K,b0,n,delta)
N(eigvec1,3) # Expressing the eigenvector in decimals upto 3rd decimal place
```

```python
# Computing the 2nd Eigenvector
# Initial guess for the 2nd Eigenvector is orthonormal to the 1st eigenvector and a unit vector
b02 = Matrix([[1/sqrt(2)],[-1/sqrt(2)]])
eigvec2,noIter2 = misesIteration((M**-1)*K,b02,n,delta)
N(eigvec2,3) # Expressing the eigenvector in decimals upto 3rd decimal place
#noIter2
```

## Central Difference Method (Explicit Scheme)

* EOM solved at $t$ to compute the displacement at $t + \Delta t$
* A special starting procedure needs to be implemented in order to compute $u_{-\Delta t}$ for $u_{1}$

```python
# Function for the central difference method
def CDM(M,C,K,F,delt,u0,u0_d,u0_dd,iters):
    # Description
    # iters: #Iterations
    # F: load vector at various time steps as a list
    # Integration Constants
    a0 = 1/(delt**2)
    a1 = 1/(2*delt)
    a2 = 2*a0
    a3 = 1/a2

    # Effective Mass Matrix
    Meff = a0*M + a1*C

    # Displacement at -delt (Special Starting Procedure)
    uss = ((delt**2)/2)*u0_dd - delt*u0_d + u0
    
    # Initialising the displacment vector at various time steps as a list
    u = []
    
    # At initial time
    u.append(uss) # at -delt
    u.append(u0) # at initial time
    
    for ii in range(1,iters+1):
        # ii coressponds to t
        # ii - 1 corresponds to t - delt
        # For force, ii - 1 corresponds to t
        # Effective Load Vector (at time = t)
        Feff = F[ii-1] - (K - a2*M)*u[ii] - (a0*M - a1*C)*u[ii-1]
        
        # Displacement vector (at time = t + delt)
        ut = (Meff**-1)*Feff
        
        # Appending the computed displacement vector to the list
        u.append(ut)
        
    return u
        
```

```python
# Number of required iterations
iters = 12
# Choice of time step
choice = '1'
# Load Vector at each time step
Ft = []
for ii in range(0,iters):
    Ft.append(F) # Constant Load Vector for the given problem

if choice == '1':
    delt = delt1
elif choice == '2':
    delt = delt2

u_cdm = CDM(M,C,K,Ft,delt,u0,u0_d,u0_dd,iters)

x1c = [] # Displacement of 1st Mass
x2c = [] # Displacement of 2nd Mass

for i in range(0,iters):
    x1c.append(u_cdm[i+2][0])
    x2c.append(u_cdm[i+2][-1])


x2c
```

```python
# Simulation time vector
t = []
t0 = 0
for j in range(0,iters):
    t.append(t0 + (j + 1)*delt)
    
    
# Plotting the displacments for the stable scheme
fig = plt.figure(figsize = (15,15))
plt.show()
plt.plot(t,x1c,label = 'Mass 1 Displacement')
plt.plot(t,x2c,label = 'Mass 2 Displacement')
plt.grid()
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Displacement (units)')
plt.title('CDM')
```

## Houbolt Integration Method (Implicit Scheme)

* EOM solved at $t + \Delta t$ to compute the displacement at $t + \Delta t$
* A special starting procedure needs to be implemented in order to compute $u_{-\Delta t}$ and $u_{-2\Delta t}$ for $u_{1}$

```python
# Function for the Houbolt Integration Scheme
def houbolt(M,C,K,F,delt,u0,u0_d,u0_dd,iters):
    # Description
    # iters: #Iterations
    # F: load vector at various time steps as a list
    # Integration Constants
    a0 = 2/(delt**2)
    a1 = 11/(6*delt)
    a2 = 5/(delt**2)
    a3 = 3/delt
    a4 = -2*a0
    a5 = -a3/2
    a6 = a0/2
    a7 = a3/9

    # Effective Stiffness Matrix
    Keff = K + a0*M + a1*C
    
    # Initialising the displacment vector at various time steps as a list
    u = []
    u.append(u0)
    
    # For the starting procedure, the exact solution for the displacment vector at t = delt is used
    u1 = Matrix([[0.003],[0.382]])
 
    
    for ii in range(1,iters+1):
        # ii coressponds to t + delt
        # ii - 1 corresponds to t 
        # ii - 2 corresponds to t - delt
        # ii - 3 corresponds to t - 2delt
        # For force, ii - 1 corresponds to t + delt
        # Effective Load Vector (at time = t + delt)
        uss1 = 2*u0 - u1 + (delt**2)*u0_dd
        uss2 = 9*u0 - 8*u1 + 6*delt*u0_d + 6*(delt**2)*u0_dd
        if ii == 1:
            Feff = F[ii-1] + M*(a2*u0 + a4*uss1 + a6*uss2) + C*(a3*u0 + a5*uss1 + a7*uss2)
        elif ii == 2:
            Feff = F[ii-1] + M*(a2*u[ii-1] + a4*u[ii-2] + a6*uss1) + C*(a3*u[ii-1] + a5*u[ii-2] + a7*uss1)
        else:
            Feff = F[ii-1] + M*(a2*u[ii-1] + a4*u[ii-2] + a6*u[ii-3]) + C*(a3*u[ii-1] + a5*u[ii-2] + a7*u[ii-3])
        
        # Displacement vector (at time = t + delt)
        ut = (Keff**-1)*Feff
        
        # Appending the computed displacement vector to the list
        u.append(ut)
        
    return u
```

```python
# Number of required iterations
iters = 12
# Choice of time step
choice = '1'
# Load Vector at each time step
Ft = []
for ii in range(0,iters+1):
    Ft.append(F) # Constant Load Vector for the given problem

if choice == '1':
    uh = houbolt(M,C,K,Ft,delt1,u0,u0_d,u0_dd,iters) # at timestep 1
elif choice == '2':
    uh = houbolt(M,C,K,Ft,delt2,u0,u0_d,u0_dd,iters) # at timestep 2

x1h = [] # Displacement of 1st Mass
x2h = [] # Displacement of 2nd Mass

for i in range(0,iters):
    x1h.append(uh[i+1][0])
    x2h.append(uh[i+1][-1])


x2h
```

## Wilson - Theta Method (Implicit Scheme)

* EOM solved at $t + \Delta t$ to compute the displacement at $t + \Delta t$
* No special starting procedure required 
* $\theta = 1.4$ for the simulation

```python
# Function for the Wilson Theta Integration Scheme
def wilson(M,C,K,F,delt,u0,u0_d,u0_dd,iters):
    # Description
    # iters: #Iterations
    # F: load vector at various time steps as a list
    theta = 1.4
    # Integration Constants
    a0 = 6/((theta*delt)**2)
    a1 = 3/(theta*delt)
    a2 = 2*a1
    a3 = (theta*delt)/2
    a4 = a0/theta
    a5 = -a2/theta
    a6 = 1 - (3/theta)
    a7 = delt/2
    a8 = (delt**2)/6

    # Effective Stiffness Matrix
    Keff = K + a0*M + a1*C
    
    # Initialising the displacment vector at various time steps as a list
    u = []
    u_d = []
    u_dd = []
    
    u.append(u0)
    u_d.append(u0_d)
    u_dd.append(u0_dd)
    
    for ii in range(1,iters+1):
        # ii coressponds to t + delt
        # ii - 1 corresponds to t 
        # ii - 2 corresponds to t - delt
        # ii - 3 corresponds to t - 2delt
        # Effective Load Vector (at time = t + theta*delt)
        Feff = F[ii-1] + theta*(F[ii] - F[ii-1]) + M*(a0*u[ii-1] + a2*u_d[ii-1] + 2*u_dd[ii-1]) + C*(a1*u[ii-1] + 2*u_d[ii-1] + a3*u_dd[ii-1])
        
        # Displacement vector (at time = t + theta*delt)
        ut = (Keff**-1)*Feff
        
        # Acceleration vector at t + delt
        ut_dd = a4*(ut-u[ii-1]) + a5*u_d[ii-1] + a6*u_dd[ii-1]
        u_dd.append(ut_dd)
        # Veloctiy at t + delt
        u_d.append(u_d[ii-1] + a7*(ut_dd + u_dd[ii-1]))
        # Displacement at t+delt
        u.append(u[ii-1] + delt*u_d[ii-1] + a8*(ut_dd + 2*u_dd[ii-1]))
        
        
    return u
```

```python
# Number of required iterations
iters = 12
# Choice of time step
choice = '2'
# Load Vector at each time step
Ft = []
for ii in range(0,iters+1):
    Ft.append(F) # Constant Load Vector for the given problem

if choice == '1':
    delt = delt1
elif choice == '2':
    delt = delt2

u_wilson = wilson(M,C,K,Ft,delt,u0,u0_d,u0_dd,iters)

x1w = [] # Displacement of 1st Mass
x2w = [] # Displacement of 2nd Mass

for i in range(0,iters):
    x1w.append(u_wilson[i+1][0])
    x2w.append(u_wilson[i+1][-1])


len(u_wilson)
x1w
#x2w
```

```python
# Simulation time vector
t = []
t0 = 0
for j in range(0,iters):
    t.append(t0 + (j + 1)*delt)
    
    
# Plotting the displacments for the stable scheme
fig = plt.figure(figsize = (15,15))
plt.show()
plt.plot(t,x1w,label = 'Mass 1 Displacement')
plt.plot(t,x2w,label = 'Mass 2 Displacement')
plt.grid()
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Displacement (units)')
plt.title('Wilson Theta')
```

## Newmark Method (Implicit Scheme)

* EOM solved at $t + \Delta t$ to compute the displacement at $t + \Delta t$
* No special starting procedure required 
* $\delta = 0.5$ and $\alpha = 0.25$ for the simulation

```python
# Function for the Newmark Scheme
def newmark(M,C,K,F,delt,u0,u0_d,u0_dd,iters):
    # Description
    # iters: #Iterations
    # F: load vector at various time steps as a list
    d = 0.5
    alp = 0.25
    # Integration Constants
    a0 = 1/(alp*(delt**2))
    a1 = d/(alp*delt)
    a2 = 1/(alp*delt)
    a3 = (1/(2*alp)) - 1
    a4 = (d/alp) - 1
    a5 = (delt/2)*((d/alp)-2)
    a6 = delt*(1-d)
    a7 = d*delt

    # Effective Stiffness Matrix
    Keff = K + a0*M + a1*C
    
    # Initialising the displacment vector at various time steps as a list
    u = []
    u_d = []
    u_dd = []
    
    u.append(u0)
    u_d.append(u0_d)
    u_dd.append(u0_dd)
    
    for ii in range(1,iters+1):
        # ii coressponds to t + delt
        # ii - 1 corresponds to t 
        # Effective Load Vector (at time = t + theta*delt)
        Feff = F[ii] + M*(a0*u[ii-1] + a2*u_d[ii-1] + a3*u_dd[ii-1]) + C*(a1*u[ii-1] + a4*u_d[ii-1] + a5*u_dd[ii-1])
        
        # Displacement vector (at time = t + delt)
        ut = (Keff**-1)*Feff
        u.append(ut)
        
        # Acceleration vector at t + delt
        ut_dd = a0*(ut-u[ii-1]) - a2*u_d[ii-1] - a3*u_dd[ii-1]
        u_dd.append(ut_dd)
        # Veloctiy at t + delt
        u_d.append(u_d[ii-1] + a6*u_dd[ii-1] + a7*ut_dd)
        
        
    return u
```

```python
# Number of required iterations
iters = 12
# Choice of time step
choice = '2'
# Load Vector at each time step
Ft = []
for ii in range(0,iters+1):
    Ft.append(F) # Constant Load Vector for the given problem

if choice == '1':
    delt = delt1
elif choice == '2':
    delt = delt2

u_newmark = newmark(M,C,K,Ft,delt,u0,u0_d,u0_dd,iters)

x1n = [] # Displacement of 1st Mass
x2n = [] # Displacement of 2nd Mass

for i in range(0,iters):
    x1n.append(u_newmark[i+1][0])
    x2n.append(u_newmark[i+1][-1])


len(u_newmark)
x1n
x2n
```

```python
# Simulation time vector
t = []
t0 = 0
for j in range(0,iters):
    t.append(t0 + (j + 1)*delt)
    
    
# Plotting the displacments for the stable scheme
fig = plt.figure(figsize = (15,15))
plt.show()
plt.plot(t,x1n,label = 'Mass 1 Displacement')
plt.plot(t,x2n,label = 'Mass 2 Displacement')
plt.grid()
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Displacement (units)')
plt.title('Newmark')
```
